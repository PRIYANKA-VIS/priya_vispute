```python
''' calculator
1. add, sub, mul, div_f, div_i, mod, exp
2. bit_or, bit_and, bit_xor
3. n_sum, n_multi, n_average
'''
import textwrap as tw


class Calculator:
    # class attr
    PI = 22/7
    
    # constructor
    def __init__(self):
        print('welcome to calculator')
        self.count = input('how many arguments do you want : ')
        
        if self.count.isnumeric(): self.count = int(self.count)
        else: print('Sorry, invalid count, please enter number.')
        
        self.arguments()
        
    def arguments(self):
        # instance attrs
        if self.count == 2:
            self.first = int(input('enter first number : '))
            self.second = int(input('enter second number : '))
            self.operator = input(tw.dedent('''
                choose from operations for two arguemnts:
                '+' : add\t'-' : sub\t'*' : mul
                '/' : div_f\t'//' : div_i\t'%' : mod\t'**' : exp
                '|' : bit_or\t'&' : bit_and\t'^' : bit_xor
                please enter your choice : ''')
            )
            self.operations()
            
        elif self.count > 2:
            self.args = tuple(map(int, input('enter numbers (space-separated) : ').split()))
            self.operator = int(input(tw.dedent('''
                choose from operations for n arguemnts:
                1. n_summation\t2. n_multiplication\t3. n_average
                please enter your choice : ''')))
            self.operations()
            
        else:  # self.count==1
            print(f'Sorry, cannot operate on {self.count} arguments')
            
    def operations(self):
        # op is variable in local scope of operation()
        
        
        self.operation = {
            '+': self.add,
            '-': self.sub,
            '*': self.mul,
            '/': self.div_f,
            '//': self.div_i,
            '%': self.mod,
            '**': self.exp,
            '|': self.bit_or,
            '&': self.bit_and,
            '^': self.bit_xor,
            1: self.n_summation,
            2: self.n_multiplication,
            3: self.n_average 
        }
        
        # calling function as per operator
        print(f'\nanswer : {self.operation[self.operator]()}')

    # instance method for two arguments
    def add(self): return self.first + self.second
    def sub(self): return self.first - self.second
    def mul(self): return self.first * self.second
    def div_f(self): return self.first / self.second
    def div_i(self): return self.first // self.second
    def mod(self): return self.first % self.second
    def exp(self): return self.first ** self.second
    def bit_or(self): return self.first | self.second
    def bit_and(self): return self.first & self.second
    def bit_xor(self): return self.first ^ self.second
    
    # instance method for arbitrary arguments
    def n_summation(self): 
        answer = 0
        for item in self.args: answer += item
        return answer     # return sum(args)
    
    def n_multiplication(self): 
        answer = 1
        for item in self.args: answer *= item
        return answer
    
    def n_average(self):
        answer = 0
        for item in self.args: answer += item
        return answer / len(self.args)
        
# using Calculator
x = Calculator()
```
